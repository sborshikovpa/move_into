### 10. Индекс слов

Дан список слов. Создай словарь, где ключ это первая буква, а значение это множество слов на эту букву.

Пример:

```python
["cat", "car", "dog"] -> {"c": {"cat", "car"}, "d": {"dog"}}
```


````python
in_list = ["cat", "car", "dog"]
def func(input:list) -> dict:
    result = {}
    for i in input:
        if i[0] in result.keys():
            result[i[0]].add(i)
        else:
            result[i[0]] = set([i])
        print(result)

func(in_list)
````

### 5. Сложение кошельков

Создай класс `Wallet` с количеством денег.

Реализуй:

- `__add__`, чтобы можно было складывать два кошелька;
- `__str__`, чтобы красиво печатать результат.


````python
class Wallet:
    def __init__(self, money):
        self._money = money
    def __add__(self, otherWallet):
        return Wallet(self._money + otherWallet._money)
    def __str__(self):
        return f"в нашем кошельке {self._money}"


print(Wallet(20) + Wallet(50))
````

### 6. Сравнение товаров

Создай класс `Product` с `name` и `price`.

Реализуй `__lt__`, чтобы товары можно было сортировать по цене через `sorted(products)`.


````python

class Product:
    def __init__(self, name, price):
        self._name = name
        self._price = price
    
    def __str__(self):
        return f"цена {self._price}"
    def __repr__(self):
        return f"цена {self._price}"  
    def __lt__(self, otherProduct):
        return self._price < otherProduct._price


glit = [Product("q",100), Product("b",80)]
glit.sort(key=lambda x: x._price)
print(glit)
    
````

### 7. Вектор

Создай класс `Vector` с `x` и `y`.

Реализуй:

- `__add__`;
- `__sub__`;
- `__str__`;
- `__eq__`.

Проверь сложение, вычитание и сравнение двух векторов.


````python
class Vector:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    def __add__(self, otherVector):
        return Vector(self._x + otherVector._x,self._y + otherVector._y)

    def __sub__(self, otherVector):
        return Vector(otherVector._x - self._x, otherVector._y - self._y)
    
    def __str__(self):
        return f"координаты {self._x}, {self._y}"
    
    def __eq__(self, otherVector):
        return self._x == otherVector._x and self._y == otherVector._y


ab = Vector(10,20)
cd = Vector(15,30)

print(ab + cd )
print(ab - cd )
print(ab == cd )
````

### 8. Коллекция задач

Создай класс `TodoList`, который хранит список задач.

Реализуй:

- `add(task)`;
- `__len__`;
- `__iter__`;
- `__str__`.

После этого объект должен работать с `len(todo)` и в цикле `for`.


````python
class TodoList:
    def __init__(self, list_of_task = None):
        
        if not list_of_task:
            self._list_of_task = []
        else:
            
            self._list_of_task = list_of_task.copy()
        
    def add(self, task):
        self._list_of_task.append(task)

    def __len__(self):
        return len(self._list_of_task)
    
    def __iter__(self):
        return iter(self._list_of_task)
    def __str__(self):
        return f"ghhg {[i for i in self._list_of_task]}"

chtoto = TodoList(["py", "exe", "zip"])

print(chtoto)
````

### 9. Фабрика уведомлений

Создай базовый класс `Notifier` с методом `send(message)`.

Создай наследников:

- `EmailNotifier`;
- `SmsNotifier`;
- `TelegramNotifier`.

Создай фабрику `NotifierFactory`, которая по строке `"email"`, `"sms"` или `"telegram"` возвращает нужный объект.


````python
class Notifier:
    def send(self, message):
        raise NotImplementedError

class EmailNotifier(Notifier):
    def send(self, message):
        print(f"Email: {message}")

class SMSNotifier(Notifier):
    def send(self, message):
        print(f"SMS: {message}")

class TelegramNotifier(Notifier):
    def send(self, message):
        print(f"Telegram: {message}")

class NotificationFactory:

    @staticmethod
    def create(notification_type):
        if notification_type == "email":
            return EmailNotifier()

        if notification_type == "sms":
            return SMSNotifier()

        if notification_type == "telegram":
            return TelegramNotifier()

        raise ValueError("Неизвестный тип уведомления")

NotificationFactory().create("email").send("ntn")
````

### 10. Фабрика игровых противников

Создай классы противников:

- `ScoutEnemy`;
- `TankEnemy`;
- `BossEnemy`.

У каждого должен быть метод `attack()`.

Создай фабрику `EnemyFactory`, которая по строке создает нужного противника. Если тип неизвестен, выбрасывай `ValueError`.


````python
class ScoutEnemy:
    def attack(self):
        return f"бьет {self.__class__.__name__}"
class TankEnemy:
    def attack(self):
        return f"бьет {self.__class__.__name__}"
class BossEnemy:
    def attack(self):
        return f"бьет {self.__class__.__name__}"

class FabricEnemy:
    @staticmethod
    def create_emeny(name_enemy):
        if name_enemy == 'ScoutEnemy':
            return ScoutEnemy()
        elif name_enemy == 'TankEnemy':
            return TankEnemy()
        elif name_enemy == 'BossEnemy':
            return BossEnemy()
        else:
            raise ValueError("Неизвестный тип")


FabricEnemy.create_emeny("ScoutEnemy").attack()
````
