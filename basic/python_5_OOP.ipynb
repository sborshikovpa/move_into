# Объектно-ориентированное программирование

Это подход к программированию, где мы описываем нашу программу не просто как набор функций, а как взаимодействие объектов.

У ООП есть три главных принципа:

1. **Инкапсуляция** — сокрытие внутренней логики объекта от внешнего мира. Мы прячем данные и предоставляем доступ к ним только через методы. С этим принципом мы уже познакомились, когда объединяли данные (например, hp, координату) и методы (перемещение, атака) внутри одного класса.
2. **Наследование** — создание новых классов на основе уже существующих, чтобы не писать один и тот же код дважды. С этим мы подробно работали в третьем уроке (когда делали класс Item и наследовали от него HealingPotion).
3. **Полиморфизм** — способность разных объектов по-разному реагировать на одни и те же команды (методы).

### Полиморфизм в Python

Полиморфизм — это один из ключевых принципов объектно-ориентированного программирования (ООП).
Слово означает «множество форм». В программировании это означает, что разные классы могут иметь методы с одинаковыми именами, но реализовывать их по-разному.

Благодаря полиморфизму мы можем использовать единый интерфейс для работы с объектами разных типов.

#### Переопределение методов (Method Overriding)

Самый частый пример полиморфизма — когда дочерний класс переопределяет метод родительского класса.

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Гав!"

class Cat(Animal):
    def speak(self):
        return "Мяу!"
```

Теперь мы можем написать функцию, которая принимает любое животное и вызывает его метод `speak`. Нам не нужно проверять, кто именно перед нами — собака или кошка.

```python
animals = [Dog(), Cat(), Dog()]

for animal in animals:
    print(animal.speak())
```

#### Утиная типизация (Duck Typing)

В Python полиморфизм тесно связан с концепцией **утиной типизации**: *«Если это выглядит как утка, плавает как утка и крякает как утка, то это, вероятно, и есть утка»*.

Python не требует, чтобы классы имели общего родителя для использования одинаковых интерфейсов. Важно только наличие нужного метода.

```python
class Bird:
    def fly(self):
        return "Птица летит"

class Airplane:
    def fly(self):
        return "Самолет летит"

def make_it_fly(flying_object):
    print(flying_object.fly())

make_it_fly(Bird())
make_it_fly(Airplane())
```

Здесь `Bird` и `Airplane` вообще никак не связаны наследованием, но функция `make_it_fly` успешно работает с обоими объектами, так как они оба имеют метод `fly`.

#### Задание

Реализуйте базовый класс `Shape` и два дочерних класса: `Rectangle` и `Circle`.

- У каждого класса должен быть метод `area()`, который возвращает площадь фигуры.
- `Rectangle` при инициализации принимает ширину и высоту.
- `Circle` при инициализации принимает радиус.
Для числа Пи используйте 3.14.

```python
class Shape:
    def area(self):
        raise Exception("Метод должен быть переопределен")


class Rectangle(Shape):
    def __init__(self, width: float, height: float):
        self._width = width
        self._height = height

    def area(self):
        return self._width * self._height


class Circle(Shape):
    def __init__(self, radius: float):
        self._radius = radius
        

    def area(self):
        return float(3.14) * self._radius ** 2

rect = Rectangle(4, 5)
circle = Circle(10)

assert rect.area() == 20
assert circle.area() == 314.0
```

### Магические методы (Dunder methods)

В Python классы имеют специальные методы, которые начинаются и заканчиваются двойным подчеркиванием `__` (отсюда название dunder - double underscore).
Они позволяют вашим объектам взаимодействовать со встроенными функциями и операторами Python (например, `+`, `==`, `print()`, `len()`).

#### `__str__` (строковое представление)

По умолчанию, если мы попробуем распечатать объект с помощью `print()`, мы увидим не очень полезную информацию (имя класса и адрес в памяти).

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p = Person("Иван", 30)
print(p)
```

Чтобы сделать вывод красивым и понятным, мы можем определить магический метод `__str__`. Он должен возвращать строку.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Человек(Имя: {self.name}, Возраст: {self.age})"

p = Person("Иван", 30)
print(p)
```

#### `__add__` (сложение)

Если мы попытаемся сложить два наших объекта через `+`, Python выдаст ошибку, так как не знает, как их складывать.
Для этого существует метод `__add__`. Давайте создадим класс `Wallet` (Кошелек) и научим его складываться с другими кошельками.

```python
class Wallet:
    def __init__(self, money):
        self.money = money

    def __add__(self, other):
        return Wallet(self.money + other.money)

    def __str__(self):
        return f"Кошелек({self.money} руб.)"

w1 = Wallet(100)
w2 = Wallet(50)
w3 = w1 + w2  # Здесь неявно вызывается w1.__add__(w2)
print(w3)
```

#### Задание

Реализуйте класс `Vector`, который представляет вектор на плоскости с координатами `x` и `y`.

Добавьте полиморфизм для оператора сложения: реализуйте метод `__add__`, чтобы два объекта `Vector` можно было складывать с помощью оператора `+`. Результатом должен быть новый объект `Vector` с суммой соответствующих координат.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(2, 3)
v2 = Vector(4, 1)
v3 = v1 + v2

assert v3.x == 6
assert v3.y == 4
```

#### Другие магические методы

**Инициализация и отображение:**
- `__init__(self, ...)` — конструктор, вызывается при создании нового объекта.
- `__str__(self)` — вызывается функцией `str()` и при `print()`. Возвращает красивую и понятную строку для пользователя.
- `__repr__(self)` — вызывается функцией `repr()`. Возвращает строку для разработчиков (удобную для отладки).

**Арифметические операции:**
- `__add__(self, other)` — сложение (`+`).
- `__sub__(self, other)` — вычитание (`-`).
- `__mul__(self, other)` — умножение (`*`).
- `__truediv__(self, other)` — деление (`/`).

**Сравнение:**
- `__eq__(self, other)` — равно (`==`).
- `__ne__(self, other)` — не равно (`!=`).
- `__lt__(self, other)` — меньше (`<`).
- `__gt__(self, other)` — больше (`>`).

**Коллекции и прочее:**
- `__len__(self)` — вызывается функцией `len()`, возвращает длину.
- `__getitem__(self, key)` — доступ по индексу или ключу (`obj[key]`).
- `__call__(self, ...)` — позволяет вызывать объект как обычную функцию (`obj()`).

#### Суть задачи

У вас есть разные модели: одна написана на библиотеке `sklearn`, другая — это нейросеть из библиотеки `transformers`, третья — вообще ваш собственный кастомный класс.

Проблема в том, что у них разные способы загрузки и получения предсказаний. У `sklearn` это метод `predict()`, у трансформеров часто нужно передавать данные напрямую в вызов объекта `model()`, и так далее.

Если мы будем писать `if / else` везде в коде, где нужно сделать предсказание, всё быстро превратится в нечитаемую кашу. Нам нужен **единый интерфейс** — чтобы мы могли просто написать `model.predict(X)`, и не думать о том, какая именно библиотека используется под капотом.

#### Паттерн «Фабрика» (Factory)

**Паттерн проектирования** — это типовое, проверенное временем решение частой проблемы в коде. «Фабрика» — это паттерн, который как раз и решает проблему создания различных объектов, предоставляя им общий интерфейс.

Идея фабрики в нашем случае:
1. Создать общий интерфейс (базовый класс) для всех моделей.
2. Для каждой конкретной библиотеки (`sklearn`, `transformers`) написать класс-обертку, который реализует этот интерфейс.
3. Сама «Фабрика» — это сущность, которая по строковому названию типа (например, `"sklearn"`) понимает, какую именно обертку нужно создать и вернуть.

Давайте посмотрим, как это выглядит на практике. Сначала мы создадим базовый класс (абстракцию) `BaseBackend`, который заставит всех своих наследников обязательно реализовать методы `load` и `predict`.

```python
from abc import ABC, abstractmethod

class BaseBackend(ABC):
    def __init__(self, path: str, device: str = "cpu"):
        self.path = path
        self.device = device
        self.native = None  # Здесь будет храниться оригинальная модель

    @abstractmethod
    def load(self):
        """Загрузить модель"""
        pass

    @abstractmethod
    def predict(self, X):
        """Сделать предсказание"""
        pass

class SklearnBackend(BaseBackend):
    def load(self):
        # В реальности здесь было бы: self.native = joblib.load(self.path)
        print(f"Загружаем sklearn модель из {self.path}...")
        self.native = "Sklearn_Model"
        return self

    def predict(self, X):
        # В реальности: return self.native.predict(X)
        return [f"Sklearn prediction for {x}" for x in X]

class TransformersBackend(BaseBackend):
    def load(self):
        # В реальности здесь была бы загрузка модели huggingface
        print(f"Загружаем transformers модель из {self.path} на устройство {self.device}...")
        self.native = "Transformers_Model"
        return self

    def predict(self, X):
        # Трансформеры обычно принимают данные иначе, здесь мы скрываем эту разницу
        return [f"Transformers prediction for {x}" for x in X]
```

Теперь напишем сам класс `Model`, который будет выступать в роли нашей «Фабрики». При создании объекта он принимает путь и тип модели, а внутри себя (под капотом) инициализирует нужный бекенд.

```python
class Model:
    def __init__(self, path: str, model_type: str, device: str = "cpu"):
        self.path = path
        self.device = device
        
        # Вот здесь работает логика фабрики:
        if model_type == "sklearn":
            self.backend = SklearnBackend(self.path, self.device).load()
        elif model_type == "transformers":
            self.backend = TransformersBackend(self.path, self.device).load()
        else:
            raise ValueError(f"Неизвестный тип модели: {model_type}")
            
    def predict(self, X):
        # Единый интерфейс для любых моделей
        return self.backend.predict(X)
```

Теперь в основном коде программы мы можем использовать разные модели **абсолютно одинаково**. 

Более того, если завтра выйдет новая библиотека (например, мы захотим добавить поддержку `XGBoost`), нам нужно будет лишь написать новый `XGBoostBackend` и добавить один `elif` в класс `Model`. Весь остальной код программы, который использует предсказания, менять не придется вообще!

```python
# Использование sklearn
model_sklearn = Model("/path/to/model.pkl", "sklearn")
y_pred_1 = model_sklearn.predict([[1.5, 2.3], [0.7, 1.8]])
print("Результат 1:", y_pred_1)

print("-" * 30)

# Использование transformers
model_transformers = Model("/path/to/transformer", "transformers", device="cuda")
y_pred_2 = model_transformers.predict(["Some text", "Another text"])
print("Результат 2:", y_pred_2)
```
