### Методы

В поле класса можно определить не только переменную, но и функцию, такая функция называется метод

Первым методом который мы определим будет `__init__()` - это специальное название для конструктора, метода, вызываемого при создании объекта

#### Инкапсуляция

Не все данные внутри класса должны быть доступны снаружи

Представьте что вы управляете самолётом, вы из кабины контролируете только штурвал, педали и кнопки, которые являются интерфейсами для управления элеронами, элевонами, закрылками, тягой и прочими прибамбахами

Сами эти прибамбахи вы не контролируете, у вас нет к ним доступа, он был бы излишним

Ограничение доступа называется инкапсуляция

В python её нет, но принято помечать поля, к которым мы не хотим чтобы имели доступ извне, нижним подчёркиванием

К таким полям всё ещё можно дотянуться извне, но потребуется явно прописать нижнее подчёркивание, и другой программист точно будет понимать что делает что-то нестандартное с вашим классом

```python
class Tree:

    def __init__(self, leaves=64, height=10.0, growth_rate=0.1):
        self._leaves = leaves
        self._height = height
        self._growth_rate = growth_rate

    def grow(self):
        self._height *= 1+self._growth_rate
        self._leaves *= 2

    def get_height(self): # на случай если потребуется обратиться к такому полю, можно реализовать вот такой метод
        return self._height

    def show(self):
        print(f"hello im DEREVO, height = {self._height}, i have {self._leaves} leaves")
```

#### Наследование

Мы уже заходим в тему которая редко используется в ML/DS, но если вы будете её знать, то очень сильно снизите частоту ситуаций когда в упор не понимаете на инструментальном уровне реализацию той или иной вещи

Я уже упоминал возможность опереть новый класс на уже созданный, это называется наследование

Класс ниже - это общий класс любого объекта на карте для нашей игры, у него прописаны x, y, char и инкапсулированы, а также есть метод проверяющий нахождение по заданным координатам

```python
class Entity:


    def __init__(self, x:int, y:int, char:str):
        self._x = x
        self._y = y
        self.char = char


    def check_cell(self, x, y):
        return self._x == x and self._y == y
```

Вот так я рассчитываю использовать, у каждой сущности будут координаты и символ, и они будут отображаться вот этой функцией, независимо от того что я там надобавляю

```python
def draw(entities):
    for y in range(5):
        for x in range(10):
            is_empty = True
            for entity in entities:
                if entity.check_cell(x, y):
                    print(entity.char, end=' ')
                    is_empty = False
            if is_empty:
                print('*', end=' ')
        print()

entities = [
    Entity(1, 2, 'p'),
]

draw(entities)
```

У каждого из моих классов должны быть эти параметры, но прописывать их всем отдельно было бы неудобно

Можно наследовать класс от другого, если при создании прописать `class класс_потомок(класс_родитель):`

Ниже я создал `Character`, он наследует у `Entity` все поля, но мы переопредилили конструктор

```python
class Character(Entity):


    def __init__(self, x:int, y:int, char:str):
        self._x = x
        self._y = y
        self.char = char
        self._max_hp = 10
        self._hp = 10
```

Метод `check_cell` не определён в `Character`, однако доступен, потому что унаследован у `Entity`

```python
character = Character(3, 4, 'c')
character.check_cell(3, 4)
```

Мы прописали в конструкторе `Character` строки, которые уже есть в конструкторе `Entity`, код не переиспользуется, это плохо

Можно вызвать метод родителя с помощью `super()`

Я им ни разу в жизни не пользовался, но про него спрашивают на собесах

```python
class Character(Entity):


    def __init__(self, x:int, y:int, char:str):
        
        # отработает конструктор Entity, где всё это задаётся
        super().__init__(x, y, char)
        
        # а затем добавим новые атрибуты
        self._max_hp = 10
        self._hp = 10


    def heal(self, hp):
        """
        заодно добавил метод лечения, интерфейс для добавления хп
        но не выше максимума!
        """
        self._hp = self._hp + hp if self._hp + hp < self._max_hp else self._max_hp
```

```python
class Item(Entity):

    def __init__(self, x:int, y:int, char:str):
        super().__init__(x, y, char)

    def effect(self, target):
        print(f"using item on {target.char}")


class Healing_salve(Item):

    def __init__(self, x:int, y:int):
        super().__init__(x, y, char='h')

    def effect(self, target):
        super().effect(target)
        target.heal(3)
```

```python
healing_potion_1 = Healing_salve(1,1)
```

```python
character._hp = 2
character._hp
```

```python
healing_potion_1.effect(target=character)
```

```python
character._hp
```

```python
entities = [
    Entity(1, 2, 'p'),
    character,
    healing_potion_1
]

draw(entities)
```
