```python
class Item:
    pass


class Weapon(Item):
    pass


class Potion(Item):
    pass


class Character:
    pass


entities = [
    Weapon(),
    Character(),
    Potion(),
    Character()
]


def filter_items(entities:list):
    result_list = []
    for i in entities:
      if isinstance(i, Item):
         result_list.append(i)
    return result_list
filtered = filter_items(entities)
print(filtered)
assert len(filtered) == 2, "В списке должно остаться ровно 3 объекта"


print("Все тесты успешно пройдены!")
```
