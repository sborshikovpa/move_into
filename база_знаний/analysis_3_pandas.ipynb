Ещё часто смотрят пропуски: метод `isna()` вернёт маску (колонку с True/False). Сам по себе этот метод бесполезен, нужно посчитать количество или долю значений True с помощью `sum()`.

### Method chaining
В pandas принят стиль кода, когда прямо для того что возвращает один метод, вызывается следующий метод, и можно записать подряд много преобразований (method chaining).


```python
# Цепочка методов:
# 1. Берем колонку order_id
# 2. Проверяем на пропуски (isna возвращает True/False)
# 3. Считаем долю пропусков (mean считает долю True)
missing_ratio = (
    df["order_id"]
    .isna()
    .mean()
)
missing_ratio
```

### Фильтрации
Чтобы выбирать из большого датафрейма строки по определённому условию, используются фильтрации. Мы пишем `df` и в квадратных скобках даём булеву маску.

Выберем заказы, где сумма заказа больше 10 000.

Двойные квадратные скобки позволяют выбрать несколько колонок по списку.


```python
big_orders = df[df["order_sum"] > 10000]
big_orders[["event_date", "order_sum", "commission", "delivery_fee"]]
```

Булева маска - это колонка того же размера, но заполненная булевым значением, которое получается по определённому условию. Всё что как-то связано с проверкой по условию в pandas, возвращает её.

```python
df["order_sum"] > 10000
```

```python
df['order_id'].isna()
```

Часто фильтруют по нескольким условиям сразу. В pandas для этого используют логические операторы: `&` (и), `|` (или). Каждое условие обязательно берём в круглые скобки.


```python
# Отбираем заказы с платной доставкой и суммой больше 10000
df_filtered = df[(df["delivery_fee"] > 0) & (df["order_sum"] > 10000)]
df_filtered[["event_date", "order_sum", "commission", "delivery_fee"]]
```

Фильтрация по списку значений делается методом `isin()`. А даты в формате строк можно сравнивать как обычные значения.


```python
paid_delivery = df[df["delivery_fee"].isin([990, 1490])]
paid_delivery["delivery_fee"].value_counts()
```

```python
df[df['event_date'] >= "2025-03-01"]
```

### Редактирование колонок
Найдём дубликаты. Метод `duplicated()` возвращает булеву маску дублей. `keep=False` помечает все повторения, включая первое, чтобы мы увидели всю группу дублей.


```python
df[df["order_id"].duplicated(keep=False)].sort_values("order_id")
```

Удалим дубликаты методом `drop_duplicates()`.


```python
# Удаляем повторяющиеся строки по order_id, оставляем только первую
df = df.drop_duplicates(subset=["order_id"], keep="first")
df
```

Для просмотра пропусков в нескольких колонках сразу можно использовать `isna().sum()`.


```python
df[["event_date", "order_sum", "commission", "delivery_fee"]].isna().sum()
```

### Преобразования таблиц
Метод `concat` нужен для объединения таблиц. `axis=0` склеивает по строкам (друг под другом), `axis=1` по столбцам.


```python
events = pd.read_csv("analysis_data/df_events.csv")

# Разделим события на две части для примера
events_early = events[events["event_date"] < "2025-02-15"]
events_late = events[events["event_date"] >= "2025-02-15"]

# Склеиваем два датафрейма по строкам
events_all = pd.concat([events_early, events_late], axis=0)
events_all
```

Загрузим табличку где каждому юзеру присвоена группа А или В

```python
matching = pd.read_csv("analysis_data/df_matching.csv")
matching
```

Метод `merge` — аналог JOIN в SQL. Параметр `on` задаёт общий ключ, а `how` тип соединения (`left`, `inner`, `right`, `outer`).


```python
orders_with_group = df.merge(
    matching[["user_id", "exp_group"]],
    on="user_id", # ключ, по которому ищем совпадения
    how="left" # оставляем все строки левой таблицы (df)
)
orders_with_group[["order_id", "user_id", "exp_group", "order_sum"]]
```

Метод `pivot_table` позволяет "повернуть" таблицу (из long формата в wide), применяя агрегационные функции.


```python
events_pivot = events.pivot_table(
    index="user_id",
    columns="event_type",
    values="event_date", # колонка, по которой считаем
    aggfunc="count", # считаем количество событий
    fill_value=0 # заполняет нулями те типы событий, которых у пользователя не было
)
events_pivot
```

### Группировки и агрегации
Сначала `groupby()` — выбираем, по чему группируем, потом `agg()` — выбираем, что именно считаем.


```python
# Группируем по экспериментальной группе и считаем несколько метрик сразу
orders_group_metrics = (
    orders_with_group
    .groupby("exp_group")
    .agg(
        orders_cnt=("order_id", "nunique"),    # Уникальные заказы
        users_cnt=("user_id", "nunique"),      # Уникальные пользователи
        avg_order_sum=("order_sum", "mean"),   # Средний чек
        total_revenue=("order_sum", "sum"),    # Общая выручка
        avg_commission=("commission", "mean"),
    )
)
orders_group_metrics
```

Для группировки по датам, чтобы увидеть динамику, используем `pd.Grouper` с указанием частоты (`freq="D"` для дней). Предварительно приведём колонку дат к типу `datetime`.


```python
# pd.to_datetime конвертирует строки в объекты дат
# assign создает или перезаписывает колонку
orders_daily = (
    df.assign(event_date=pd.to_datetime(df["event_date"]))
    .groupby(pd.Grouper(key="event_date", freq="D"))
    .agg(
        orders_cnt=("order_id", "nunique"),
        revenue=("order_sum", "sum"),
        avg_check=("order_sum", "mean"),
    )
)
orders_daily
```
